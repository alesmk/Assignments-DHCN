```python
import pandas as pd
import matplotlib.pyplot as plt

# Leggi il file output.log
file_path = 'output.log'

# Leggiamo i dati dal file
data = pd.read_csv(file_path, header=None, names=['t_par'])

# Aggiungiamo la colonna con il numero di thread (1 a 50)
data['threads'] = range(1, len(data) + 1)

# Creiamo il grafico
plt.plot(data['threads'], data['t_par'], marker='o')

# Etichettiamo gli assi
plt.xlabel('Number of workers')
plt.ylabel('T_c')

# Titolo del grafico
#plt.title('Tempo di completamento rispetto al numero di thread')

# Mostra il grafico
plt.grid(True)
plt.show()
```
